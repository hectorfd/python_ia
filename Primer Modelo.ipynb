# Primer Modelo Regresion Lineal

```python
import pandas as pd
# Para abrir un archivo CSV 
datos = pd.read_csv('celsius.csv')# nuestro archivo tiene como nombre celsius.csv
```

```python
# Para obtener o viasualizar los datos usaremos la funcion .info()
datos.info()# esto nos mostrara el # de columna, la cabecera, el tipo de dato, cantidad e incluso el tipo de dato
```

```python
# Para ver un vistazo de los datos usamos head()
datos.head()
```

```python
# Para visualizar los datos usaremos seaborn
import seaborn as sb
# debes de escribir las cabeceras tal cual sino te dara error
# hacemos uso de las coordendas x e y
# data es una variable que contendra los datos y recuerda que datos contiene el archivo celsius.csv
# hue -> es para darle color en referencia a la columna fahreneit
# palette -> es para darle una paleta de colores coolwarm es adecaudo porque da colores frios a calidos
sb.scatterplot(x='celsius', y='fahrenheit', data=datos, hue='fahrenheit', palette='coolwarm')
```

## Primero Preparar los Datos
Separamos los datos es caracteristicas(x), etiquetas(y)

```python
# creamos las variables x y y como vimos arriba x = caracteristicas, y = etiquetas
x = datos['celsius']
y = datos['fahrenheit']
```

```python
# Con solo escribir la variable que declaramos arriba mostrara los datos de esa columna
y
```

```python
# los datos actualmente son de tipo arreglo
#! x.values # = array([-40, -10,   0,   8,  15,  22,  38])
# los datos para trabajarlos deben estar cada uno en un arreglo unico # ([[-40],[-10],[0],[8],...])
# entonces para esa transformacion de numpy llamada reshape
x_procesada = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

## Segundo Paso Crear el Modelo

```python
#sklearn es una libreria de machine learning 
# importamos LinearRegression que nos permitira crear el modelo
from sklearn.linear_model import LinearRegression
modelo = LinearRegression() # aun no podemos usarlo porque ahora sigue el entrenamiento
```

## Tercer Paso Entrenar 

```python
# como tenemos pocos datos en este ejemplo esto sera sencillo
modelo.fit(x_procesada, y_procesada)# fit es para entrenar el modelo con los datos preparados
```

## Probar la prediccion con datos de prueba

```python
# predeciremos el valor de fahrenheit para celcius 8 que es 46.4 fahrenheit
modelo.predict([[8]])
```

```python
# como 8 viene en el cojunto de datos, podemos dar cualquier numero y luego comprobarlo en google
modelo.predict([[222]])# 222 celcius es 431.60 fahrenheit
```

```python
# usaresmo score para que nos evalue el modelo dandonos un valor entre el 0 y el 1 mientras mas cerca
# del 1 mejor sera el modelo, este modelo nos dara 1 porque es perfecto
modelo.score(x_procesada, y_procesada)
```

Este es un ejemplo sencillo de hacer ya que la relacion es perfectamente lineal el modelo predecira de manera exacta
